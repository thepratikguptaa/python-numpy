## Advance Operations with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data Structures: [restaurant_id, 2022, 2023, 2024, 2025]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Biryani
    [2, 120000, 140000, 160000, 190000], # Beijing Bites
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Barn
    [5, 160000, 185000, 205000, 230000]  # Chai Point
])
```

```python
print("==== Zomato sales analysis ====")
print("\n Sales data shape: ", sales_data.shape)
print("\n Sample data for first 3 restaurants:", sales_data[:3])
print("\n Sample data for all restaurants but without the IDs:", sales_data[:, 1:])
```

```python
# total sales per year
print("\n Total sales per year:", np.sum(sales_data[:, 1:], axis=0))
# total sales per restaurant
print("\n Total sales per restaurant:", np.sum(sales_data[:, 1:], axis=1))

# Minimum sales per retaurant
min_sales = np.min(sales_data[:, 1:], axis=1)
print("\n Minimum sales per restaurant:", min_sales)
# Maximum sales per retaurant
max_sales = np.max(sales_data[:, 1:], axis=1)
print("\n Maximum sales per restaurant:", max_sales)
# Average sales per retaurant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("\n Average sales per restaurant:", avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("\n Cumulative sum per restaurant:", cumsum)

plt.figure(figsize=(8, 4))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales per year across all restaurants")
plt.xlabel("Year")
plt.ylabel("Cumulative sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)

print("\n Multiplication vector: ", vector1 * vector2)

print("\n Dot product: ", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\n Angle between vectors: ", angle)
```

```python
restaurant_types = np.array(["Italian", "Mexican", "Chinese", "Indian", "Japanese"])
print("\n Restaurant types: ", restaurant_types)
print("\n In Lower case: ", np.vectorize(str.lower)(restaurant_types))
print("\n In Upper case: ", np.vectorize(str.upper)(restaurant_types))
```
